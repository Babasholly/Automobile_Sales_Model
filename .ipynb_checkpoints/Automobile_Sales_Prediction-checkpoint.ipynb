#### Using AI generate data dictionary

- ORDERNUMBER: The unique identifier for each order.
- QUANTITYORDERED: The quantity of products ordered in each order.
- PRICEEACH: The price per unit of the product ordered.
- ORDERLINENUMBER: The line number for each item in the order.
- SALES: The total sales amount for each order.
- ORDERDATE: The date when the order was placed.
- DAYS_SINCE_LASTORDER: The number of days since the last order was placed by the customer.
- STATUS: The status of the order (e.g., pending, shipped, delivered).
- PRODUCTLINE: The category or type of the product.
- MSRP: Manufacturer's Suggested Retail Price for the product.
- PRODUCTCODE: The unique code identifying each product.
- CUSTOMERNAME: The name of the customer placing the order.
- PHONE: The contact phone number of the customer.
- ADDRESSLINE1: The address line 1 of the customer.
- CITY: The city where the customer is located.
- POSTALCODE: The postal code of the customer's address.
- COUNTRY: The country where the customer is located.
- CONTACTLASTNAME: The last name of the contact person for the order.
- CONTACTFIRSTNAME: The first name of the contact person for the order.
- DEALSIZE: The size or scale of the deal (e.g., small, medium, large)

#### Visual Analysis of the Categorical features

```python

import plotly.express as px

cat_columns = as_df[['STATUS', 'PRODUCTLINE', 'DEALSIZE']]
def Analysis_category(cols): 
    print("Distribution of", cols)  # checking the individual visual from buy bar and pie chart 
    print("_"* 60)
    colors = ['#79a5db', '#e0a580', '#6fab90', '#896ca8', '#ADD8E6']
    value_counts = cat_columns[cols].value_counts()
    
    # Count plot
    fig = px.bar(value_counts,x=value_counts.index, y=value_counts.values, title=f'Distribution of {cols}',
        labels={'x': 'Categories', 'y': 'Count'},color_discrete_sequence=[colors])
    fig.update_layout(width=700)
    fig.update_layout(plot_bgcolor='#ffffff', paper_bgcolor='#ffffff')
    fig.show()
    
    # Donut chart
    percentage = (value_counts / value_counts.sum()) * 100
    fig = px.pie( values=percentage, names=value_counts.index,
        labels={'names': 'Categories', 'values': 'Percentage'}, hole=0.5,color_discrete_sequence=colors)
    fig.add_annotation( x=0.5, y=0.5, align="center", xref="paper",
        yref="paper", showarrow=False, font_size=15, text=f'{cols}')
    fig.update_layout(legend=dict(x=0.9, y=0.5))
    fig.update_layout(width=700)
    fig.show()
    
    print("       ")
for x in cat_columns:
    Analysis_category(x)

    
```

#### Visual Analysing the Numerical feature

```python
def univariateAnalysis_numeric(column, nbins):
    print("Description of " + column)
    print("----------------------------------------------------------------------------")
    print(as_df[column].describe(), end=' ')
    fig, axes = plt.subplots(1, 2, figsize=(15, 6))
    
    # Plot histogram
    print("----------------------------------------------------------------------------")
    sns.histplot(as_df[column], bins=nbins, kde=True, ax=axes[0])
    axes[0].set_title("Histogram of " + column)
    
    # Plot violin plot
    print("----------------------------------------------------------------------------")
    sns.boxplot(x=as_df[column], ax=axes[1])
    axes[1].set_title("Boxlot of " + column )
    plt.show()
for x in df_num:
    univariateAnalysis_numeric(x,20)
```

```python
X = as_df.drop ('STATUS', axis = 1)
y = as_df.loc['STATUS']
```

```python
from sklearn.feature_selection import SelectKBest, chi2

X = as_df[['PRICEEACH','PRODUCTLINE', 'DEALSIZE']]
y = as_df['SALES']
```

```python
as_df['SALES'].value_counts()
```

```python
X.head()
```

```python
X.tail()
```

```python
X['PRODUCTLINE'] = X['PRODUCTLINE'].replace({'Ships':0, 'Motorcycles':1, 'Planes':2, 'Classic Cars':3, 'Trains':4, 'Vintage Cars':5, 'Trucks and Buses':6})
```

```python
X['DEALSIZE'] = X['DEALSIZE'].replace({'Small':0, 'Medium':1, 'Large':2})
```

```python
X.head()
```

```python
X.tail()
```

```python
as_df['SALES'].value_counts()
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_X = scaler.fit_transform(X)
```

```python
X_new = SelectKBest(k=3).fit_transform(X, y)

X_new
```

```python
## Split the dataset 

from sklearn.model_selection import train_test_split

# Split data 70%-30% into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)
```

```python
# Train the model
from sklearn.linear_model import LinearRegression

# Fit a linear regression model on the training set
model = LinearRegression().fit(X_train, y_train)
print (model)
```

```python
## Evaluate the trained model

from sklearn.metrics import mean_squared_error, r2_score

predictions = model.predict(X_test)

mse = mean_squared_error(y_test, predictions)
print("MSE:", mse)

rmse = np.sqrt(mse)
print("RMSE:", rmse)

r2 = r2_score(y_test, predictions)
print("R2:", r2)
```

```python
## check if there's an overfitting

r2_score(y_train, model.predict(X_train))
```

The result from the shown that the model 

#### Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import export_text

# Train the model
model = DecisionTreeRegressor().fit(X, y)
print (model, "\n")

# Visualize the model tree
tree = export_text(model)
print(tree)
```

```python
# Evaluate the model using the test data
predictions = model.predict(X_test)
mse = mean_squared_error(y_test, predictions)
print("MSE:", mse)
rmse = np.sqrt(mse)
print("RMSE:", rmse)
r2 = r2_score(y_test, predictions)
print("R2:", r2)

# Plot predicted vs actual
plt.scatter(y_test, predictions)
plt.xlabel('Actual Labels')
plt.ylabel('Predicted Labels')
plt.title('Automobile Sales Predictions')
# overlay the regression line
z = np.polyfit(y_test, predictions, 1)
p = np.poly1d(z)
plt.plot(y_test,p(y_test), color='magenta')
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor

# Train the model
model = RandomForestRegressor().fit(X_train, y_train)
print (model, "\n")

# Evaluate the model using the test data
predictions = model.predict(X_train)
mse = mean_squared_error(y_train, predictions)
print("MSE:", mse)
rmse = np.sqrt(mse)
print("RMSE:", rmse)
r2 = r2_score(y_train, predictions)
print("R2:", r2)


# Plot predicted vs actual
plt.scatter(y_train, predictions)
plt.xlabel('Actual Labels')
plt.ylabel('Predicted Labels')
plt.title('Automobile Sale Predictions')
# overlay the regression line
z = np.polyfit(y_train, predictions, 1)
p = np.poly1d(z)
plt.plot(y_train,p(y_train), color='magenta')
plt.show()
```

#### Findings 

From the analysis above, it was observed that the R2 of the train data is higher when compared to the test  data. This result shows that the model is overfitted and suitable for prediction.
